```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
import pandas as pd

# Create a small sample dataset
data = {
    "Cement": [540, 540, 332, 332, 198],
    "Slag": [0, 0, 142, 142, 132],
    "FlyAsh": [0, 0, 0, 0, 0],
    "Water": [162, 162, 228, 228, 192],
    "Superplasticizer": [2.5, 2.5, 0, 0, 0],
    "CoarseAggregate": [1040, 1055, 932, 932, 978],
    "FineAggregate": [676, 676, 594, 594, 825],
    "Age": [28, 28, 270, 365, 360],
    "Strength": [79.99, 61.89, 40.27, 41.05, 44.30]  # target column
}

concrete_data = pd.DataFrame(data)

# Split into features (X) and target (y)
X = concrete_data.drop("Strength", axis=1)
y = concrete_data["Strength"]

# Train Linear Regression on the full dataset
lr_model = LinearRegression()
lr_model.fit(X, y)
lr_predictions = lr_model.predict(X)

# Train Random Forest on the full dataset
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X, y)
rf_predictions = rf_model.predict(X)

# Evaluate on the same dataset (not real validation!)
print(f"Linear Regression R² Score: {r2_score(y, lr_predictions):.2f}")
print(f"Random Forest R² Score: {r2_score(y, rf_predictions):.2f}")
```
