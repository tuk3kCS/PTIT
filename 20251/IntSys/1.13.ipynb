```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Dataset (Height in meters, Weight in kg)
# Đây là bộ dữ liệu hợp lý hơn (tham khảo từ sách ML):
heights = np.array([[1.60], [1.65], [1.70], [1.73], [1.80]])
weights = np.array([[60], [65], [72.3], [75], [80]])

print("\nOriginal Data (Height in meters, Weight in kg):")
for i in range(len(heights)):
    print(f"Height: {heights[i, 0]:.2f}m, Weight: {weights[i, 0]:.1f} kg")

# 2. Train Linear Regression model
model = LinearRegression()
model.fit(heights, weights)

# 3. Get coefficient and intercept (as scalars, not arrays)
gradient = model.coef_.item()
intercept = model.intercept_.item()
print(f"\nLearned Regression Line: Weight = {gradient:.2f} * Height + {intercept:.2f}")

# 4. Define new heights to predict
new_heights = np.array([[1.62], [1.78], [1.85]])
predicted_weights = model.predict(new_heights)

print("\nPredicted Weights for Specific New Heights:")
for i in range(len(new_heights)):
    print(f"For a height of {new_heights[i, 0]:.2f}m, predicted weight = {predicted_weights[i, 0]:.2f} kg")

# 5. Visualization
plt.figure(figsize=(10, 6))
plt.scatter(heights, weights, color='blue', label='Original Data', s=100)
plt.plot(heights, model.predict(heights), color='red', label='Regression Line')
plt.scatter(new_heights, predicted_weights, color='green', marker='X', s=150, label='Predictions', zorder=5)

plt.title('Weight Prediction from Height (Linear Regression)')
plt.xlabel('Height (m)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.legend()
plt.axis([1.55, 1.90, 50, 100])  # adjust axis range for clarity
plt.show()
```
