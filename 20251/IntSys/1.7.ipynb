```python
# Nguyễn Hoàng Tùng

import numpy as np

# Define the arrays for names, heights, and weights as specified in the assignment [2]
names = np.array(['Ann','Joe','Mark'])
heights = np.array([1.5, 1.78, 1.6]) # Heights in meters [2]
weights = np.array([65, 46, 59])    # Weights in kgs [2]

# Calculate the Body Mass Index (BMI) using the formula: BMI = weight / (height^2) [1, 2]
# NumPy allows for element-wise operations on arrays, making this calculation straightforward [1]
bmi = weights / heights ** 2

# Print the calculated BMI for each individual [1]
print("Calculated BMI values:", bmi)

# Further, you can use Boolean indexing to identify individuals based on their BMI,
# as demonstrated in the source [1]:
print("\nBMI Categories:")
print("Overweight: " , names[bmi > 25])                  # Individuals with BMI > 25 [1]
print("Underweight: " , names[bmi < 18.5])               # Individuals with BMI < 18.5 [1]
print("Healthy: " , names[(bmi >= 18.5) & (bmi <= 25)])    # Individuals with BMI between 18.5 and 25 (inclusive) [1]
```
