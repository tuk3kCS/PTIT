```python
# Nguyễn Hoàng Tùng

# ==============================================
# CNN cơ bản để phân loại ảnh MNIST (TensorFlow)
# ==============================================

import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

# 1️⃣ Tải và tiền xử lý dữ liệu
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()

# Chuyển thành ảnh 28x28x1 (1 kênh màu xám)
x_train = x_train.reshape((x_train.shape[0], 28, 28, 1)).astype('float32') / 255
x_test = x_test.reshape((x_test.shape[0], 28, 28, 1)).astype('float32') / 255

# 2️⃣ Xây dựng kiến trúc CNN
model = models.Sequential([
    # Lớp tích chập 1: 32 bộ lọc 3x3
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),

    # Lớp tích chập 2: 64 bộ lọc 3x3
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Lớp tích chập 3: 64 bộ lọc 3x3
    layers.Conv2D(64, (3, 3), activation='relu'),

    # Flatten để đưa sang fully connected
    layers.Flatten(),

    # Fully connected 64 neuron
    layers.Dense(64, activation='relu'),

    # Lớp output 10 lớp tương ứng 10 chữ số
    layers.Dense(10, activation='softmax')
])

# 3️⃣ Biên dịch mô hình
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 4️⃣ Huấn luyện mô hình
history = model.fit(x_train, y_train, epochs=5,
                    validation_data=(x_test, y_test))

# 5️⃣ Đánh giá mô hình
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"\n✅ Độ chính xác trên tập kiểm tra: {test_acc*100:.2f}%")

# 6️⃣ (Tuỳ chọn) Vẽ biểu đồ huấn luyện
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Nguyễn Hoàng Tùng

# ==============================================
# RNN cơ bản (SimpleRNN) phân loại chữ số MNIST
# ==============================================

import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

# 1️⃣ Tải dữ liệu MNIST
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()

# Chuẩn hóa dữ liệu về [0,1]
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# 2️⃣ Xây dựng mô hình RNN
model = models.Sequential([
    # Mỗi ảnh 28x28 được xem như chuỗi 28 bước, mỗi bước 28 giá trị
    layers.SimpleRNN(128, activation='tanh', input_shape=(28, 28)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# 3️⃣ Biên dịch mô hình
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 4️⃣ Huấn luyện mô hình
history = model.fit(x_train, y_train, epochs=5,
                    validation_data=(x_test, y_test))

# 5️⃣ Đánh giá mô hình
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"\n✅ Độ chính xác trên tập kiểm tra: {test_acc*100:.2f}%")

# 6️⃣ (Tuỳ chọn) Vẽ biểu đồ độ chính xác
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Nguyễn Hoàng Tùng

# =====================================================
# LSTM cơ bản - Dự đoán giá trị tiếp theo trong chuỗi sin
# =====================================================

import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# 1️⃣ Tạo dữ liệu chuỗi sin
x = np.linspace(0, 100, 1000)
y = np.sin(x)

# Chuẩn bị dữ liệu huấn luyện: chia chuỗi thành các đoạn nhỏ (window)
time_steps = 20

def create_dataset(series, time_steps=1):
    X, Y = [], []
    for i in range(len(series) - time_steps):
        X.append(series[i:(i + time_steps)])
        Y.append(series[i + time_steps])
    return np.array(X), np.array(Y)

X, Y = create_dataset(y, time_steps)

# Thêm 1 chiều cho LSTM: (samples, time_steps, features)
X = X.reshape((X.shape[0], X.shape[1], 1))

# 2️⃣ Chia tập huấn luyện & kiểm thử
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
Y_train, Y_test = Y[:split], Y[split:]

# 3️⃣ Xây dựng mô hình LSTM
model = Sequential([
    LSTM(64, activation='tanh', input_shape=(time_steps, 1)),
    Dense(32, activation='relu'),
    Dense(1)
])

# 4️⃣ Biên dịch và huấn luyện
model.compile(optimizer='adam', loss='mse')
history = model.fit(X_train, Y_train, epochs=20, batch_size=16,
                    validation_data=(X_test, Y_test), verbose=1)

# 5️⃣ Dự đoán
Y_pred = model.predict(X_test)

# 6️⃣ Vẽ kết quả
plt.figure(figsize=(10,5))
plt.plot(Y_test, label='Giá trị thực tế', linewidth=2)
plt.plot(Y_pred, label='Dự đoán (LSTM)', linestyle='--')
plt.legend()
plt.title("Dự báo chuỗi sin bằng LSTM")
plt.xlabel("Thời gian")
plt.ylabel("Giá trị sin")
plt.show()
```

```python
# Nguyễn Hoàng Tùng

# Notebook: RNN & LSTM time-series forecasting for Hanoi, Danang, HCM
# Run in Jupyter. Make sure tensorflow is installed: pip install tensorflow

import os
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---------- CONFIG ----------
# Thay đổi đường dẫn nếu cần (nếu chạy local Windows, bạn có thể set DATA_DIR = r"D:\PTIT\20251\IntSys")
DATA_DIRS = [Path("/mnt/data"), Path(r"D:\PTIT\20251\IntSys")]
# Tên file (nếu khác, sửa ở đây)
FILES = ["df_weather.csv", "precipitation_mean_max_min_temp_1901_2020_en.csv", "weather.csv"]

TARGET_CITIES = ["Hà Nội", "Đà Nẵng", "Hồ Chí Minh"]  # tên hiển thị
LOOKBACK = 14  # dùng 14 ngày trước để dự đoán 1 ngày tới
BATCH_SIZE = 32
MAX_EPOCHS = 30

# ---------- try import tensorflow ----------
try:
    import tensorflow as tf
    from tensorflow import keras
    from tensorflow.keras import layers
except Exception as e:
    raise RuntimeError("TensorFlow không được tìm thấy. Cài bằng `pip install tensorflow` rồi chạy lại.") from e

# ---------- helper: find and load CSVs ----------
def find_files():
    found = {}
    for d in DATA_DIRS:
        if not d.exists():
            continue
        for fname in FILES:
            p = d / fname
            if p.exists():
                found[fname] = p
    return found

found = find_files()
if not found:
    print("Không tìm thấy file trong các đường dẫn mặc định. Kiểm tra DATA_DIRS / FILES.")
else:
    print("Tìm thấy:", found)

# Load available files into dataframes
dfs = {}
for fname, path in found.items():
    try:
        df = pd.read_csv(path)
    except Exception:
        df = pd.read_csv(path, encoding="latin1")
    dfs[fname] = df
    print(f"\nLoaded {fname} shape={df.shape}")
    print(df.columns.tolist()[:20])

# ---------- heuristic: extract per-city temperature series ----------
# We'll attempt to locate columns: date, city, temperature-like.
def find_temp_col(cols):
    cols_low = [c.lower() for c in cols]
    # prefer column containing 'temp' but not 'min'/'max' (for avg)
    for i,c in enumerate(cols_low):
        if "temp" in c and "max" not in c and "min" not in c:
            return cols[i]
    # fallback common names
    for cand in ("temperature", "tavg", "t_mean", "t"):
        for i,c in enumerate(cols_low):
            if cand in c:
                return cols[i]
    return None

city_series = {}  # map display_name -> pd.Series(index=date, values=temp)

# Try each loaded df to find city/time/temp
for name, df in dfs.items():
    cols = df.columns.tolist()
    date_col = None
    city_col = None
    # find date column
    for c in cols:
        if "date" in c.lower() or "time" in c.lower() or "day" in c.lower():
            date_col = c
            break
    # find city column
    for c in cols:
        if any(x in c.lower() for x in ("city","station","location","name")):
            city_col = c
            break
    temp_col = find_temp_col(cols)
    # If file is pivoted (column per city), detect and use directly
    if not city_col and any(any(tc.lower() in c.lower() for tc in ["hanoi","ha noi","hà nội","ha noi","da nang","dà nẵng","ho chi minh","hcm","hochiminh","saigon"]) for c in cols):
        # find columns that match cities and create series (assume index is date or range)
        if date_col:
            df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
            for city in TARGET_CITIES:
                # try match by substring
                for c in cols:
                    if city.split()[0].lower()[:4] in c.lower():
                        ser = df.set_index(date_col)[c].dropna().sort_index()
                        city_series[city] = ser.astype(float)
        else:
            # no date; skip this approach
            pass
    elif date_col and city_col and temp_col:
        df2 = df[[date_col, city_col, temp_col]].copy()
        df2[date_col] = pd.to_datetime(df2[date_col], errors="coerce")
        df2 = df2.dropna(subset=[date_col,city_col,temp_col])
        for city in TARGET_CITIES:
            mask = df2[city_col].astype(str).str.lower().str.contains(city.split()[0].lower()[:4])
            if mask.any():
                ser = df2.loc[mask].set_index(date_col)[temp_col].sort_index()
                city_series[city] = ser.astype(float)

# If still missing, try scanning columns names in all dfs for city substrings
if not city_series:
    for name, df in dfs.items():
        for c in df.columns:
            for city in TARGET_CITIES:
                if city.split()[0].lower()[:4] in str(c).lower():
                    # try to use column as series; need index
                    if 'date' in (col.lower() for col in df.columns):
                        datecol = [col for col in df.columns if 'date' in col.lower()][0]
                        s = pd.Series(df[c].values, index=pd.to_datetime(df[datecol], errors="coerce"))
                        city_series[city] = s.dropna().astype(float)

if not city_series:
    raise RuntimeError("Không tự động trích xuất được chuỗi nhiệt độ cho các thành phố. Bạn hãy chỉ rõ cột date/city/temp hoặc cung cấp file per-city.")

print("\nCities found for modeling:", list(city_series.keys()))

# ---------- prepare supervised dataset from series ----------
def prepare_series(series, lookback=LOOKBACK):
    s = series.dropna().sort_index()
    # Nếu có duplicate index (nhiều giá trị cùng ngày), lấy trung bình theo ngày
    if isinstance(s.index, pd.DatetimeIndex):
        if not s.index.is_unique:
            s = s.groupby(s.index).mean()
        # reindex về tần suất ngày, điền giá trị còn thiếu
        s = s.asfreq('D')
        s = s.interpolate(limit_direction='both')
    arr = s.values.astype(np.float32)
    X, y = [], []
    for i in range(lookback, len(arr)):
        X.append(arr[i - lookback:i])
        y.append(arr[i])  # dự đoán ngày tiếp theo
    X = np.array(X)[:, :, None]  # shape (samples, lookback, 1)
    y = np.array(y)
    return X, y, s

def split_dataset(X, y, train_pct=0.7, val_pct=0.15):
    n = len(X)
    n_train = int(n * train_pct)
    n_val = int(n * (train_pct + val_pct))
    return (X[:n_train], y[:n_train]), (X[n_train:n_val], y[n_train:n_val]), (X[n_val:], y[n_val:])

# ---------- model builders ----------
def build_simple_rnn(input_shape):
    inp = keras.Input(shape=input_shape)
    x = layers.SimpleRNN(32)(inp)
    x = layers.Dense(8, activation='relu')(x)
    out = layers.Dense(1)(x)
    model = keras.Model(inp, out)
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

def build_lstm(input_shape):
    inp = keras.Input(shape=input_shape)
    x = layers.LSTM(32)(inp)
    x = layers.Dense(8, activation='relu')(x)
    out = layers.Dense(1)(x)
    model = keras.Model(inp, out)
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

# ---------- training loop per city ----------
results = {}
models_dir = Path.cwd() / "weather_models"
models_dir.mkdir(exist_ok=True)

for city, series in city_series.items():
    print(f"\n--- Processing city: {city} ---")
    X, y, full_series = prepare_series(series, LOOKBACK)
    if len(X) < 80:
        print(f"  (Dữ liệu ít: only {len(X)} samples) — recommend collect more history. Skipping training for {city}.")
        continue
    (X_train,y_train),(X_val,y_val),(X_test,y_test) = split_dataset(X,y)
    # normalize based on train
    mean = X_train.mean()
    std = X_train.std()
    X_train_n = (X_train - mean) / (std + 1e-6)
    X_val_n = (X_val - mean) / (std + 1e-6)
    X_test_n = (X_test - mean) / (std + 1e-6)
    y_train_n = (y_train - mean) / (std + 1e-6)
    y_val_n = (y_val - mean) / (std + 1e-6)
    y_test_n = (y_test - mean) / (std + 1e-6)
    
    # build models
    rnn = build_simple_rnn(X_train_n.shape[1:])
    lstm = build_lstm(X_train_n.shape[1:])
    # callbacks
    cb = [keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)]
    # train (kept small to run quickly)
    print("  Training SimpleRNN...")
    rnn_hist = rnn.fit(X_train_n, y_train_n, validation_data=(X_val_n, y_val_n),
                       epochs=MAX_EPOCHS, batch_size=BATCH_SIZE, callbacks=cb, verbose=1)
    print("  Training LSTM...")
    lstm_hist = lstm.fit(X_train_n, y_train_n, validation_data=(X_val_n, y_val_n),
                         epochs=MAX_EPOCHS, batch_size=BATCH_SIZE, callbacks=cb, verbose=1)
    # evaluate
    rnn_eval = rnn.evaluate(X_test_n, y_test_n, verbose=0)
    lstm_eval = lstm.evaluate(X_test_n, y_test_n, verbose=0)
    print(f"  RNN test (mse, mae): {rnn_eval}")
    print(f"  LSTM test (mse, mae): {lstm_eval}")
    # predictions for last 30 test samples
    nplot = min(30, len(X_test_n))
    preds_rnn_n = rnn.predict(X_test_n[-nplot:]).squeeze()
    preds_lstm_n = lstm.predict(X_test_n[-nplot:]).squeeze()
    preds_rnn = preds_rnn_n * (std + 1e-6) + mean
    preds_lstm = preds_lstm_n * (std + 1e-6) + mean
    true = y_test[-nplot:]
    # dates for those test samples if available
    dates = None
    if isinstance(full_series.index, pd.DatetimeIndex):
        total_samples = len(full_series) - LOOKBACK
        test_start = int(len(X_train) + len(X_val))
        idx_from = LOOKBACK + test_start
        dates = full_series.index[idx_from: idx_from + len(X_test)][-nplot:]
    # plot
    plt.figure(figsize=(9,4))
    plt.plot(true, label='true')
    plt.plot(preds_rnn, label='rnn_pred')
    plt.plot(preds_lstm, label='lstm_pred')
    plt.title(f"{city} — last {nplot} test days (true vs preds)")
    plt.legend()
    plt.tight_layout()
    plt.show()
    
    # save models
    rnn_path = models_dir / f"{city.replace(' ', '_')}_rnn.keras"
    lstm_path = models_dir / f"{city.replace(' ', '_')}_lstm.keras"
    rnn.save(rnn_path)
    lstm.save(lstm_path)
    
    results[city] = {
        "rnn_eval": rnn_eval,
        "lstm_eval": lstm_eval,
        "rnn_model_path": str(rnn_path),
        "lstm_model_path": str(lstm_path)
    }

# summary
if results:
    df_summary = pd.DataFrame([
        {
            "city": c,
            "rnn_mse": r["rnn_eval"][0],
            "rnn_mae": r["rnn_eval"][1],
            "lstm_mse": r["lstm_eval"][0],
            "lstm_mae": r["lstm_eval"][1],
            "rnn_model": r["rnn_model_path"],
            "lstm_model": r["lstm_model_path"]
        }
        for c,r in results.items()
    ])
    print("\n=== Training summary ===")
    display(df_summary)
    print("Models saved under:", models_dir)
else:
    print("No city processed (dữ liệu không đủ hoặc không tìm thấy series).")
```

```python
# Nguyễn Hoàng Tùng

import os
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error
import tensorflow as tf
from tensorflow import keras

# ========================
# ⚙️ Cấu hình
# ========================
DATA_DIR = Path(r"D:\PTIT\20251\IntSys")  # Thư mục chứa 3 file
MODEL_DIR = Path("weather_models")         # Thư mục chứa các model .keras
LOOKBACK = 14                              # Số ngày để dự đoán ngày kế
TARGET_CITIES = ["Hà Nội", "Đà Nẵng", "Hồ Chí Minh"]

# ========================
# 🧩 1. Nạp dữ liệu
# ========================
print("Đang tải dữ liệu...")

df_weather = pd.read_csv(DATA_DIR / "df_weather.csv")
weather = pd.read_csv(DATA_DIR / "weather.csv")
precip = pd.read_csv(DATA_DIR / "precipitation_mean_max_min_temp_1901_2020_en.csv")

# Chuyển kiểu ngày
for df in [df_weather, weather]:
    if "date" in df.columns:
        df["date"] = pd.to_datetime(df["date"], errors="coerce")

print("✅ Đã nạp dữ liệu:")
print(" - df_weather:", df_weather.shape)
print(" - weather:", weather.shape)
print(" - precip:", precip.shape)

# ========================
# 🔍 2. Trích xuất dữ liệu nhiệt độ trung bình cho 3 thành phố
# ========================

def extract_city_series(city):
    city_lower = city.lower()

    # --- ƯU TIÊN DỮ LIỆU TỪ weather.csv ---
    if "province" in weather.columns:
        mask = weather["province"].astype(str).str.contains(
            "Hà Nội|Ha Noi|Đà Nẵng|Da Nang|TP.? ?HCM|TP.? ?Hồ Chí Minh|Ho Chi Minh|HoChiMinh|Sai Gon|Saigon",
            case=False, na=False, regex=True
        )
        if mask.any():
            # lọc đúng theo tên thành phố
            if "hà" in city_lower:
                mask_city = weather["province"].astype(str).str.contains("Hà Nội|Ha Noi", case=False, na=False, regex=True)
            elif "đà" in city_lower or "da" in city_lower:
                mask_city = weather["province"].astype(str).str.contains("Đà Nẵng|Da Nang", case=False, na=False, regex=True)
            elif "hồ" in city_lower or "chi" in city_lower or "hcm" in city_lower or "sài" in city_lower:
                mask_city = weather["province"].astype(str).str.contains(
                    "Hồ Chí Minh|TP.? ?HCM|TP.? ?Hồ Chí Minh|Ho Chi Minh|HoChiMinh|Sai Gon|Saigon",
                    case=False, na=False, regex=True
                )
            else:
                mask_city = pd.Series(False, index=weather.index)

            df_city = weather.loc[mask_city].copy()
            if len(df_city) > 0:
                df_city = df_city.dropna(subset=["max", "min", "date"])
                df_city["temp_avg"] = (df_city["max"] + df_city["min"]) / 2.0
                df_city = df_city.groupby("date")["temp_avg"].mean().sort_index()
                return df_city

    # --- DỮ LIỆU DỰ PHÒNG TỪ df_weather.csv ---
    if "hà" in city_lower:
        mask = df_weather["location.name"].astype(str).str.contains("Hà Nội", case=False, na=False)
        series = df_weather.loc[mask].set_index("date")["day.avgtemp_c"]
    elif "đà" in city_lower or "da" in city_lower:
        mask = df_weather["location.name"].astype(str).str.contains("Đà Nẵng", case=False, na=False)
        series = df_weather.loc[mask].set_index("date")["day.avgtemp_c"]
    elif "hồ" in city_lower or "chi" in city_lower or "hcm" in city_lower:
        mask = df_weather["location.name"].astype(str).str.contains("Hồ Chí Minh|Ho Chi Minh", case=False, na=False)
        series = df_weather.loc[mask].set_index("date")["day.avgtemp_c"]
    else:
        series = pd.Series(dtype=float)

    series = series.dropna().sort_index()
    if not series.index.is_unique:
        series = series.groupby(series.index).mean()
    return series


# Tạo dữ liệu cho các thành phố
city_series = {city: extract_city_series(city) for city in TARGET_CITIES}

for c, s in city_series.items():
    print(f"{c}: {len(s)} bản ghi, {s.index.min()} → {s.index.max() if len(s)>0 else 'N/A'}")

# ========================
# ⚙️ 3. Tạo dataset
# ========================
def make_dataset(series, lookback=LOOKBACK):
    s = series.copy().dropna().sort_index()
    if isinstance(s.index, pd.DatetimeIndex):
        s = s.asfreq('D')
        s = s.interpolate(limit_direction='both')
    arr = s.values.astype(np.float32)
    X, y = [], []
    for i in range(lookback, len(arr)):
        X.append(arr[i-lookback:i])
        y.append(arr[i])
    X = np.array(X)[:, :, None]
    y = np.array(y)
    return X, y, s

# ========================
# 🧠 4. Load mô hình
# ========================
model_files = list(MODEL_DIR.glob("*.keras"))
if not model_files:
    raise FileNotFoundError(f"Không tìm thấy model trong {MODEL_DIR}")

loaded_models = {}
for file in model_files:
    model = keras.models.load_model(file)
    model_type = "RNN" if "rnn" in file.stem.lower() else "LSTM"
    city_key = None
    for city in TARGET_CITIES:
        if city.split()[0].lower() in file.stem.lower():
            city_key = city
            break
    if city_key:
        loaded_models.setdefault(city_key, {})[model_type] = model

print("\n✅ Đã load mô hình:")
for c, models in loaded_models.items():
    print(f" - {c}: {list(models.keys())}")

# ========================
# 📈 5. So sánh mô hình
# ========================
results = []

for city in TARGET_CITIES:
    if city not in city_series or len(city_series[city]) < 60:
        print(f"⚠️ {city}: không đủ dữ liệu để đánh giá.")
        continue

    series = city_series[city]
    X, y, s = make_dataset(series)
    mean, std = X.mean(), X.std()
    Xn = (X - mean) / (std + 1e-6)

    if city not in loaded_models:
        print(f"⚠️ Không tìm thấy model cho {city}")
        continue

    for model_name, model in loaded_models[city].items():
        preds = model.predict(Xn).squeeze() * (std + 1e-6) + mean
        mae = mean_absolute_error(y, preds)
        mse = mean_squared_error(y, preds)
        results.append({"Thành phố": city, "Mô hình": model_name, "MAE": mae, "MSE": mse})

        plt.figure(figsize=(9,4))
        plt.plot(y[-30:], label="Thực tế", color="black")
        plt.plot(preds[-30:], label=f"{model_name} dự báo", linestyle="--")
        plt.title(f"So sánh dự báo và thực tế – {city} ({model_name})")
        plt.xlabel("Ngày (cuối)")
        plt.ylabel("Nhiệt độ (°C)")
        plt.legend()
        plt.grid(alpha=0.3)
        plt.tight_layout()
        plt.show()

# ========================
# 📊 6. Tổng hợp kết quả
# ========================
if not results:
    raise RuntimeError("Không có kết quả để vẽ. Kiểm tra dữ liệu và model.")

df_results = pd.DataFrame(results)
print("\nBảng kết quả:")
print(df_results)

plt.figure(figsize=(8,5))
sns.barplot(data=df_results, x="Thành phố", y="MAE", hue="Mô hình")
plt.title("So sánh sai số MAE giữa các mô hình")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(data=df_results, x="Thành phố", y="MSE", hue="Mô hình")
plt.title("So sánh sai số MSE giữa các mô hình")
plt.tight_layout()
plt.show()

out_path = DATA_DIR / "compare_results.xlsx"
df_results.to_excel(out_path, index=False)
print(f"\n✅ Kết quả đã lưu: {out_path}")
```

```python
# Nguyễn Hoàng Tùng

import nest_asyncio
nest_asyncio.apply()

import threading
import uvicorn
from fastapi import FastAPI
from fastapi.responses import HTMLResponse, JSONResponse
from IPython.display import IFrame, display
from pathlib import Path
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import mean_absolute_error, mean_squared_error

# --------------------------
# Config
# --------------------------
DATA_DIR = Path(r"D:\PTIT\20251\IntSys")   # adjust if needed
MODEL_DIR = Path("weather_models")          # models folder (must contain .keras files)
LOOKBACK = 14
PORT = 8000

# Cities and coords
CITIES = {
    "HaNoi": {"name": "Hà Nội", "lat": 21.0285, "lon": 105.8542},
    "DaNang": {"name": "Đà Nẵng", "lat": 16.0544, "lon": 108.2022},
    "HCM": {"name": "Hồ Chí Minh", "lat": 10.8231, "lon": 106.6297},
}

# Observed temps for today (you can change these or pass via /?ha_noi=..)
DEFAULT_OBS = {"HaNoi": 25.0, "DaNang": 27.0, "HCM": 28.0}
TODAY_DATE = pd.Timestamp("2025-10-30")  # per your message

# --------------------------
# Load datafiles (safe)
# --------------------------
df_weather = pd.DataFrame()
weather = pd.DataFrame()
precip = pd.DataFrame()

if (DATA_DIR / "df_weather.csv").exists():
    try:
        df_weather = pd.read_csv(DATA_DIR / "df_weather.csv")
        if "date" in df_weather.columns:
            df_weather["date"] = pd.to_datetime(df_weather["date"], errors="coerce")
    except Exception as e:
        print("Failed to load df_weather.csv:", e)

if (DATA_DIR / "weather.csv").exists():
    try:
        weather = pd.read_csv(DATA_DIR / "weather.csv")
        if "date" in weather.columns:
            weather["date"] = pd.to_datetime(weather["date"], errors="coerce")
    except Exception as e:
        print("Failed to load weather.csv:", e)

if (DATA_DIR / "precipitation_mean_max_min_temp_1901_2020_en.csv").exists():
    try:
        precip = pd.read_csv(DATA_DIR / "precipitation_mean_max_min_temp_1901_2020_en.csv")
    except Exception as e:
        print("Failed to load precip file:", e)

print("Loaded shapes -> df_weather:", getattr(df_weather,'shape',None),
      "weather:", getattr(weather,'shape',None),
      "precip:", getattr(precip,'shape',None))

# --------------------------
def extract_city_series(city_key):
    city_patterns = {
        "HaNoi": r"Hà Nội|Ha Noi|HANOI",
        "DaNang": r"Đà Nẵng|Da Nang|DANANG",
        "HCM": r"Hồ Chí Minh|Ho Chi Minh|HoChiMinh|HCM|Sai Gon|Saigon|TP.? ?HCM"
    }
    pat = city_patterns.get(city_key)
    # try weather first
    if not weather.empty and "province" in weather.columns:
        try:
            mask = weather["province"].astype(str).str.contains(pat, case=False, na=False, regex=True)
            dfc = weather.loc[mask].dropna(subset=["date"])
            if "max" in dfc.columns and "min" in dfc.columns:
                dfc = dfc.dropna(subset=["max", "min"])
                dfc = dfc.copy()
                dfc["temp_avg"] = (dfc["max"].astype(float) + dfc["min"].astype(float)) / 2.0
                series = dfc.groupby("date")["temp_avg"].mean().sort_index()
                if len(series) > 0:
                    # ensure datetime index
                    if not isinstance(series.index, pd.DatetimeIndex):
                        try:
                            series.index = pd.to_datetime(series.index)
                        except Exception:
                            pass
                    return series
        except Exception as e:
            print("Warning: extract from weather.csv failed for", city_key, e)

    # fallback to df_weather
    if not df_weather.empty:
        try:
            # df_weather has column 'location.name' and 'day.avgtemp_c' per earlier printout
            if "location.name" in df_weather.columns and "day.avgtemp_c" in df_weather.columns:
                mask = df_weather["location.name"].astype(str).str.contains(pat.split("|")[0], case=False, na=False)
                dfc = df_weather.loc[mask].dropna(subset=["date","day.avgtemp_c"])
                if len(dfc) > 0:
                    series = dfc.set_index("date")["day.avgtemp_c"].astype(float).sort_index()
                    if not series.index.is_unique:
                        series = series.groupby(series.index).mean()
                    return series
        except Exception as e:
            print("Warning: extract from df_weather failed for", city_key, e)

    return pd.Series(dtype=float)

# --------------------------
# Safe dataset maker
# --------------------------
def make_dataset(series, lookback=LOOKBACK):
    s = series.copy().dropna().sort_index()
    if isinstance(s.index, pd.DatetimeIndex):
        s = s.asfreq('D')
        s = s.interpolate(limit_direction='both')
    arr = s.values.astype(np.float32)
    if len(arr) <= lookback:
        return np.empty((0, lookback, 1)), np.empty((0,))
    X, y = [], []
    for i in range(lookback, len(arr)):
        X.append(arr[i-lookback:i])
        y.append(arr[i])
    X = np.array(X)[:, :, None]
    y = np.array(y)
    return X, y

# --------------------------
# Load saved models (.keras) robustly
# --------------------------
MODELS = {}  # {city_key: {"RNN": model, "LSTM": model}}
if MODEL_DIR.exists():
    for p in MODEL_DIR.glob("*.keras"):
        name = p.stem.lower()
        try:
            m = keras.models.load_model(p)
        except Exception as e:
            print("Failed to load model", p.name, ":", e)
            continue
        city = None
        if ("ha" in name and "noi" in name) or "hanoi" in name:
            city = "HaNoi"
        elif "da" in name and "nang" in name or "danang" in name:
            city = "DaNang"
        elif "hcm" in name or "hochiminh" in name or "ho_chi_minh" in name or "saigon" in name or "sai gon" in name:
            city = "HCM"
        if not city:
            # fallback try matching full names
            for k,v in CITIES.items():
                if v["name"].split()[0].lower() in name:
                    city = k
                    break
        if not city:
            print("Skipping unmatched model file:", p.name)
            continue
        kind = "RNN" if "rnn" in name else ("LSTM" if "lstm" in name else "MODEL")
        MODELS.setdefault(city, {})[kind] = m
        print(f"Loaded {kind} -> {city} ({p.name})")
else:
    print("Model dir not found:", MODEL_DIR)

# --------------------------
# Forecast function (safe)
# --------------------------
def forecast_city_from_models(city_key, obs_today):
    series = extract_city_series(city_key)
    if series.empty:
        return {"error":"no_history"}
    # append today's observation (use provided TODAY_DATE)
    s = series.copy()
    s.loc[TODAY_DATE] = obs_today
    s = s.sort_index()
    if isinstance(s.index, pd.DatetimeIndex) and not s.index.is_unique:
        s = s.groupby(s.index).mean()
    X, y = make_dataset(s)
    if X.size == 0:
        return {"error":"not_enough_data"}
    # normalize based on X
    mean = X.mean()
    std = X.std() if X.std() != 0 else 1.0
    last = X[-1:]
    last_n = (last - mean) / (std + 1e-6)
    out = {}
    for kind, model in MODELS.get(city_key, {}).items():
        try:
            p = model.predict(last_n, verbose=0).squeeze()
            p_real = float(p * (std + 1e-6) + mean)
            out[kind] = round(p_real,2)
        except Exception as e:
            out[kind] = None
            print("Prediction error", city_key, kind, e)
    # optionally compute naive baseline (persist last observed)
    out["baseline_last_obs"] = float(s.iloc[-1])
    return out

# --------------------------
# Build results for all cities (use DEFAULT_OBS or query param)
# --------------------------
def build_map_payload(obs_override=None):
    obs = DEFAULT_OBS.copy()
    if isinstance(obs_override, dict):
        for k,v in obs_override.items():
            if k in obs:
                obs[k] = float(v)
    payload = {"today": TODAY_DATE.strftime("%Y-%m-%d"), "cities": {}}
    for key,meta in CITIES.items():
        preds = forecast_city_from_models(key, obs.get(key, None))
        payload["cities"][key] = {
            "name": meta["name"],
            "lat": meta["lat"],
            "lon": meta["lon"],
            "observed_today": obs.get(key),
            "predictions": preds
        }
    return payload

# --------------------------
# FastAPI app
# --------------------------
app = FastAPI()

@app.get("/", response_class=HTMLResponse)
def index():
    # embed payload into HTML/JS
    payload = build_map_payload()
    html = f"""
    <!doctype html>
    <html>
    <head>
      <meta charset="utf-8"/>
      <title>Weather Forecast Map</title>
      <meta name="viewport" content="width=device-width, initial-scale=1.0">
      <link rel="stylesheet" href="https://unpkg.com/leaflet@1.9.4/dist/leaflet.css" />
      <style>
        #map{{height:90vh;width:100%}} body{{margin:0;font-family:Arial,Helvetica,sans-serif}}
        .popup-table td{{padding:3px 6px}}
      </style>
    </head>
    <body>
      <div id="map"></div>
      <script src="https://unpkg.com/leaflet@1.9.4/dist/leaflet.js"></script>
      <script>
        const payload = {payload};
        const map = L.map('map').setView([16.2,107.0],6);
        L.tileLayer('https://{{s}}.tile.openstreetmap.org/{{z}}/{{x}}/{{y}}.png', {{
            maxZoom: 19, attribution: '© OpenStreetMap'
        }}).addTo(map);
        for (const key of Object.keys(payload.cities)) {{
            const c = payload.cities[key];
            let preds = c.predictions;
            let html = `<div style="font-size:14px"><strong>${{c.name}}</strong><table class="popup-table">`;
            html += `<tr><td>Quan sát (ngày ${payload.today}):</td><td style="text-align:right"><strong>${{c.observed_today}} °C</strong></td></tr>`;
            if (preds.error) {{
                html += `<tr><td colspan="2" style="color:#b33">Không đủ dữ liệu lịch sử để dự báo</td></tr>`;
            }} else {{
                for (const k of Object.keys(preds)) {{
                    if (k==='baseline_last_obs') continue;
                    html += `<tr><td>Dự báo (${k}):</td><td style="text-align:right"><strong>${{preds[k]===null? 'N/A' : preds[k]+' °C'}}</strong></td></tr>`;
                }}
                html += `<tr><td>Baseline (last obs):</td><td style="text-align:right">${{preds.baseline_last_obs}} °C</td></tr>`;
            }}
            html += `</table></div>`;
            L.marker([c.lat,c.lon]).addTo(map).bindPopup(html);
        }}
      </script>
    </body>
    </html>
    """
    return HTMLResponse(html)

@app.get("/api/payload")
def api_payload(ha_noi: float = None, da_nang: float = None, hcm: float = None):
    # allow overriding observed values by query params
    override = {}
    if ha_noi is not None: override["HaNoi"] = ha_noi
    if da_nang is not None: override["DaNang"] = da_nang
    if hcm is not None: override["HCM"] = hcm
    return JSONResponse(build_map_payload(override))

# --------------------------
# Run server in background and display IFrame
# --------------------------
def run():
    uvicorn.run(app, host="127.0.0.1", port=PORT, log_level="error")

thread = threading.Thread(target=run, daemon=True)
thread.start()

# show in notebook
display(IFrame(f"http://127.0.0.1:{PORT}", width="100%", height=600))
print("Server started at http://127.0.0.1:%d — if iframe blank, open that URL in browser." % PORT)
```
