```python
# Cell 1: Import thư viện cần thiết
import os
import pandas as pd

# Cell 2: Xác định đường dẫn có tồn tại
paths = [
    r"D:\lecture\PTIT\20251\IntSys\Viet Nam Stock Market Data (VNINDEX, VN30, VN100)",
    r"D:\PTIT\20251\IntSys\Viet Nam Stock Market Data (VNINDEX, VN30, VN100)"
]

data_path = None
for path in paths:
    if os.path.exists(path):
        data_path = path
        break

if data_path is None:
    raise FileNotFoundError("Không tìm thấy thư mục dữ liệu chứng khoán trên cả hai đường dẫn.")

print(f"Đường dẫn dữ liệu được sử dụng: {data_path}")

# Cell 3: Load 3 dataset VNINDEX, VN30, VN100
files = {
    "VNINDEX": "HOSEVNINDEXM30.csv",
    "VN30": "HOSEVN30M30.csv",
    "VN100": "HOSEVN100M30.csv"
}

datasets = {}
for name, file in files.items():
    file_path = os.path.join(data_path, file)
    if os.path.exists(file_path):
        datasets[name] = pd.read_csv(file_path)
        print(f"{name} loaded successfully! Shape: {datasets[name].shape}")
    else:
        print(f"Không tìm thấy file: {file_path}")

# Cell 4: Kiểm tra dữ liệu đầu tiên
for name, df in datasets.items():
    print(f"\n{name} head:")
    display(df.head())
```
