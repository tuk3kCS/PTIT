```python
# Nguyễn Hoàng Tùng

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd # Used for better tabular data display

# 1. Generate 4 sets of student names, 5 subject names, and grades on a scale of 4
student_names = np.array(['Nguyễn Hoàng Tùng', 'Đinh Quyết Thắng', 'Trần Thành Trung', 'Tường Duy Minh Tú'])
subjects = np.array([
    'Introduction to Software Engineering',
    'Introduction to Artificial Intelligent',
    'Web Development',
    'Database Management System',
    'Internship'
])
num_students = len(student_names)
num_subjects = len(subjects)

# Generate random grades (scale of 0.0 to 4.0) for each student in each subject
# np.random.uniform generates random floats within a specified range.
# grades array will have shape (num_students, num_subjects)
grades = np.random.uniform(0.0, 4.0, size=(num_students, num_subjects))

# Calculate "Intelligent System Development" score (average of the 5 subjects) for each student
# This serves as the consolidated result named 'Intelligent System Development'.
# The 'axis=1' argument means the mean is calculated across the second dimension (subjects).
intelligent_system_development_scores = np.mean(grades, axis=1)

# Display the generated tabular data for clarity
print("\nGenerated Student Grades (on a scale of 4.0):")
grades_df = pd.DataFrame(grades, index=student_names, columns=subjects)
print(grades_df.round(2)) # Display grades rounded to 2 decimal places

print("\nIntelligent System Development Scores (Average Grade):")
for i in range(num_students):
    print(f"- {student_names[i]}: {intelligent_system_development_scores[i]:.2f}")

# 2. Display forms corresponding to page 71-73 of book 1

# --- Plotting Multiple Lines in the Same Chart (pages 71-72, book 1) --- [1, 2]
# This chart shows each student's performance across the 5 subjects.
plt.figure(figsize=(12, 7)) # Adjust figure size for better readability

# Loop through each student to plot their grades as a separate line
for i in range(num_students):
    # plt.plot() is used to plot a line chart [1]
    # The 'label' argument is used for the legend [2]
    plt.plot(subjects, grades[i], marker='o', linestyle='-', label=student_names[i])

# Adding a title and labels for the axes [4]
plt.title('Student Performance Across Different Subjects', fontsize=14)
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Grade (Scale of 4.0)', fontsize=12)
plt.ylim(0, 4.2) # Set y-axis limits to clearly show grades on a scale of 4.0
plt.xticks(rotation=45, ha='right', fontsize=10) # Rotate x-axis labels for readability [5]
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.6) # Add a grid for better data interpretation
plt.legend(title='Student', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10, title_fontsize=12) # Adding a legend to distinguish lines [2, 6]
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()

# --- Plotting Bar Charts (pages 73-77, book 1) --- [3]
# This chart compares the "Intelligent System Development" score (average grade) for each student.
plt.figure(figsize=(10, 6)) # Adjust figure size
bar_width = 0.6 # Width of the bars

# plt.bar() is used to plot a bar chart [3]
bars = plt.bar(student_names, intelligent_system_development_scores, width=bar_width, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])

# Adding a title and labels for the axes [4]
plt.title('Intelligent System Development Scores per Student', fontsize=14)
plt.xlabel('Student Name', fontsize=12)
plt.ylabel('Average Grade (Scale of 4.0)', fontsize=12)
plt.ylim(0, 4.2) # Set y-axis limits to match the grade scale
plt.xticks(rotation=45, ha='right', fontsize=10) # Rotate x-axis labels if needed
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.6) # Add horizontal grid lines for readability [7]

# Add value labels on top of the bars for clarity (not explicitly on page 73 but enhances understanding)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, round(yval, 2), ha='center', va='bottom', fontsize=9)

plt.tight_layout() # Adjust layout
plt.show()
```
