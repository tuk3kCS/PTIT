```python
import os
import cv2
import albumentations as A
from tqdm import tqdm

# ====== Cấu hình ======
input_dir = r"C:\DATA\anh1"       # thư mục ảnh gốc
output_dir = os.path.join(input_dir, "augmented")  # thư mục lưu ảnh augment
os.makedirs(output_dir, exist_ok=True)

# ====== 5 kỹ thuật augmentation ======
transform = A.Compose([
    A.HorizontalFlip(p=0.5),          # 1. Lật ngang
    A.RandomBrightnessContrast(p=0.5),# 2. Thay đổi độ sáng / tương phản
    A.Rotate(limit=25, p=0.5),        # 3. Xoay ±25 độ
    A.RandomCrop(width=200, height=200, p=0.5), # 4. Cắt ngẫu nhiên
    A.GaussianBlur(blur_limit=(3, 5), p=0.3)    # 5. Làm mờ Gaussian
])

# ====== 3️⃣ Hàm sinh ảnh ======
image_files = [f for f in os.listdir(input_dir)
               if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

count = 0
target_count = 20  # sinh 20 ảnh

for img_name in tqdm(image_files, desc="Augmenting"):
    img_path = os.path.join(input_dir, img_name)
    image = cv2.imread(img_path)

    if image is None:
        continue

    # Lặp cho đến khi đạt đủ 20 ảnh
    for i in range(20):  # sinh 4 phiên bản/ảnh gốc (có thể tùy chỉnh)
        augmented = transform(image=image)['image']
        save_name = f"aug_{count:02d}_{img_name}"
        save_path = os.path.join(output_dir, save_name)
        cv2.imwrite(save_path, augmented)
        count += 1

        if count >= target_count:
            break
    if count >= target_count:
        break

print(f"✅ Đã tạo {count} ảnh augment trong: {output_dir}")
```
