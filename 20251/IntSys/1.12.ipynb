```python
# Nguyễn Hoàng Tùng

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Data (diện tích m², giá nhà tỷ)
X = np.array([50, 60, 65, 70, 75, 80, 85]).reshape(-1, 1)
y = np.array([2.5, 3, 3.5, 3.8, 4, 4.5, 5])

# 2. Fit Linear Regression model
model = LinearRegression()
model.fit(X, y)

# 3. Prediction for given sizes
X_pred = np.array([55, 68, 76, 90]).reshape(-1, 1)
y_pred = model.predict(X_pred)

# Print results
for size, price in zip(X_pred.flatten(), y_pred):
    print(f"House size {size} m² → Predicted price: {price:.2f} tỷ")

# 4. Visualization
# Scatter original data
plt.scatter(X, y, color='blue', label="Original data")

# Regression line
x_line = np.linspace(45, 95, 100).reshape(-1, 1)
y_line = model.predict(x_line)
plt.plot(x_line, y_line, color='red', label="Regression line")

# Prediction points
plt.scatter(X_pred, y_pred, color='green', marker='x', s=100, label="Predictions")

# Labels and title
plt.xlabel("Diện tích (m²)")
plt.ylabel("Giá nhà (tỷ đồng)")
plt.title("Linear Regression: House Price Prediction")
plt.legend()
plt.grid(True)
plt.show()
```
