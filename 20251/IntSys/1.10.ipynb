```python
# Nguyễn Hoàng Tùng

import pandas as pd         # Used for manipulating tabular data [2, 3]
import numpy as np          # Used for numerical operations, especially array creation and random number generation [4]
import seaborn as sns       # Used for statistical data visualization, built on matplotlib [5, 6]
import matplotlib.pyplot as plt # Used for plotting charts and figures [5, 7]

# Define the local path to the Titanic dataset
# As specified in the query, the data is expected to be at D:\PTIT\IntSys\titanic.csv
local_file_path = r"D:\PTIT\IntSys\titanic.csv"

# 1. Load the Titanic dataset from the local CSV file
# Pandas provides the read_csv() function to load data from CSV files into a DataFrame [3, 8]
try:
    titanic = pd.read_csv(local_file_path)
    print(titanic.head()) # Display the first few rows of the DataFrame [9]
    print(f"\nTitanic dataset loaded with {len(titanic)} samples.")

    # 2. Plot a chart to show the survival rate using Seaborn's categorical plot
    # The task explicitly refers to page 86 of book 1 for `sns.catplot` [10] and
    # page 91 of book 2 provides a direct example for the Titanic dataset [1].
    g = sns.catplot(
        x="who",          # Column for the main categories on the x-axis: 'who' represents 'man', 'woman', 'child'
        y="survived",     # Numerical variable for which to calculate the proportion (mean): 'survived' (0 or 1)
        col="class",      # Column to create separate subplots (columns of plots) for each class
        data=titanic,     # The DataFrame containing the data
        kind="bar",       # Type of plot: 'bar' for mean/proportion visualization [11]
        ci=None,          # Do not show confidence intervals, as in the example on page 91 [11]
        aspect=1,         # Aspect ratio of each facet, similar to the example [11]
        palette="viridis" # Color palette for the bars (you can choose other palettes like "crest", "Blues", etc.)
    )

    # Customize the plot for clarity as shown in the source [1, 12]
    g.set_axis_labels("", "Survival Rate") # Set y-axis label, x-axis label is implicitly "who"
    g.set_xticklabels(["Men", "Women", "Children"]) # Set x-tick labels for 'who' categories
    g.set_titles("{col_name} {col_var}") # Set titles for each subplot, using 'class' as column variable
    g.fig.suptitle("Survival Rate of Men, Women, and Children by Class on the Titanic", y=1.02, fontsize=16) # Set a main title for the entire figure

    plt.tight_layout() # Adjust layout to prevent labels from overlapping
    plt.show()         # Display the plot

except FileNotFoundError:
    print(f"Error: The file '{local_file_path}' was not found. Please ensure the path is correct and the file exists.")
except pd.errors.EmptyDataError:
    print(f"Error: The file '{local_file_path}' is empty.")
except pd.errors.ParserError:
    print(f"Error: Could not parse the file '{local_file_path}'. Please check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
