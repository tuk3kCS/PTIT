```python
# Nguyễn Hoàng Tùng

# importing the dependencies
import tensorflow.compat.v1 as tf
import numpy as np
import matplotlib.pyplot as plt
# Model Parameters
learning_rate = 0.01
training_epochs = 2000
display_step = 200
# Training Data
train_X = np.asarray([3.3,4.4,5.5,6.71,6.93,4.168,9.779,6.182,7.59,2.167,7.042,10.791,5.313,7.997,5.654,9.27,3.1])
train_y = np.asarray([1.7,2.76,2.09,3.19,1.694,1.573,3.366,2.596,2.53,1.221,2.827,3.465,1.65,2.904,2.42,2.94,1.3])
n_samples = train_X.shape[0]
# Test Data
test_X = np.asarray([6.83, 4.668, 8.9, 7.91, 5.7, 8.7, 3.1, 2.1])
# Set placeholders for feature and target vectors
X = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
# Set model weights and bias)
test_y = np.asarray([1.84, 2.273, 3.2, 2.831, 2.92, 3.24, 1.35, 1.03])
W = tf.Variable(np.random.randn(), name="weight")
b = tf.Variable(np.random.randn(), name="bias")
# Construct a linear model
linear_model = W*X + b
# Mean squared error
cost = tf.reduce_sum(tf.square(linear_model - y)) / (2*n_samples)
# Gradient descent
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
# Initializing the variables
init = tf.global_variables_initializer()
# Launch the graph
with tf.Session() as sess:
    # Load initialized variables in current session
    sess.run(init)
    # Fit all training data
    for epoch in range(training_epochs):
        # perform gradient descent step
        sess.run(optimizer, feed_dict={X: train_X, y: train_y})

        # Display logs per epoch step
        if (epoch+1) % display_step == 0:
            c = sess.run(cost, feed_dict={X: train_X, y: train_y})
            print("Epoch:{0:6} \t Cost:{1:10.4} \t W:{2:6.4} \t b:{3:6.4}".
            format(epoch+1, c, sess.run(W), sess.run(b)))

    # Print final parameter values
    print("Optimization Finished!")
    training_cost = sess.run(cost, feed_dict={X: train_X, y: train_y})
    print("Final training cost:", training_cost, "W:", sess.run(W), "b:", sess.run(b), '\n')

    # Graphic display
    plt.plot(train_X, train_y, 'ro', label='Original data')
    plt.plot(train_X, sess.run(W) * train_X + sess.run(b), label='Fitted line')
    plt.legend()
    plt.show()

    # Testing the model
    testing_cost = sess.run(tf.reduce_sum(tf.square(linear_model - y)) / (2 * test_X.shape[0]),
    feed_dict={X: test_X, y: test_y})

    print("Final testing cost:", testing_cost)
    print("Absolute mean square loss difference:", abs(training_cost - testing_cost))
    # Display fitted line on test data
    plt.plot(test_X, test_y, 'bo', label='Testing data')
    plt.plot(train_X, sess.run(W) * train_X + sess.run(b), label='Fitted line')
    plt.legend()
    plt.show()
```

```python
# Nguyễn Hoàng Tùng

import numpy as np
import tensorflow.compat.v1 as tf
tf.disable_eager_execution()   # make sure we are in TF1.x graph mode

# ------------------------------
# NumPy Implementation
# ------------------------------
def neuron_numpy(X, W, b, activation="tanh"):
    # Weighted sum
    z = np.dot(X, W) + b
    
    # Apply activation
    if activation == "tanh":
        return np.tanh(z)
    elif activation == "sigmoid":
        return 1 / (1 + np.exp(-z))
    elif activation == "relu":
        return np.maximum(0, z)
    else:
        return z  # linear (no activation)


# ------------------------------
# TensorFlow v1 Implementation
# ------------------------------
def neuron_tf(X, W, b, activation="tanh"):
    # Define constants
    X_tf = tf.constant(X, dtype=tf.float32)
    W_tf = tf.constant(W, dtype=tf.float32)
    b_tf = tf.constant(b, dtype=tf.float32)

    # Weighted sum
    z = tf.reduce_sum(X_tf * W_tf) + b_tf

    # Choose activation
    if activation == "tanh":
        act = tf.math.tanh(z)
    elif activation == "sigmoid":
        act = tf.math.sigmoid(z)
    elif activation == "relu":
        act = tf.nn.relu(z)
    else:
        act = z  # linear

    # Run inside a session
    with tf.Session() as sess:
        return sess.run(act)


# ------------------------------
# Example usage
# ------------------------------
X = [2.0, 3.0, 4.0]
W = [0.05, 0.07, 0.09]
b = 0.60

print("=== NumPy Implementation ===")
print("tanh:", neuron_numpy(X, W, b, "tanh"))
print("sigmoid:", neuron_numpy(X, W, b, "sigmoid"))
print("relu:", neuron_numpy(X, W, b, "relu"))
print("linear:", neuron_numpy(X, W, b, None))

print("\n=== TensorFlow v1 Implementation ===")
print("tanh:", neuron_tf(X, W, b, "tanh"))
print("sigmoid:", neuron_tf(X, W, b, "sigmoid"))
print("relu:", neuron_tf(X, W, b, "relu"))
print("linear:", neuron_tf(X, W, b, None))
```

```python
# Nguyễn Hoàng Tùng

import tensorflow.compat.v1 as tf
import numpy as np
tf.disable_eager_execution()   # Use TF1.x style

# ---------------------------
# Define network parameters
# ---------------------------
n_inputs = 3       # Age, Hours of study, Previous scores
n_hidden1 = 5      # neurons in first hidden layer
n_hidden2 = 5      # neurons in second hidden layer
n_outputs = 2      # Pass / Fail

learning_rate = 0.01
training_epochs = 1000

# ---------------------------
# Placeholders for input/output
# ---------------------------
X = tf.placeholder(tf.float32, [None, n_inputs])   # input features
y = tf.placeholder(tf.float32, [None, n_outputs])  # target labels (one-hot)

# ---------------------------
# Weights & Biases
# ---------------------------
weights = {
    'h1': tf.Variable(tf.random_normal([n_inputs, n_hidden1])),
    'h2': tf.Variable(tf.random_normal([n_hidden1, n_hidden2])),
    'out': tf.Variable(tf.random_normal([n_hidden2, n_outputs]))
}
biases = {
    'b1': tf.Variable(tf.zeros([n_hidden1])),
    'b2': tf.Variable(tf.zeros([n_hidden2])),
    'out': tf.Variable(tf.zeros([n_outputs]))
}

# ---------------------------
# Build Model
# ---------------------------
# Layer 1
layer1 = tf.add(tf.matmul(X, weights['h1']), biases['b1'])
layer1 = tf.nn.relu(layer1)

# Layer 2
layer2 = tf.add(tf.matmul(layer1, weights['h2']), biases['b2'])
layer2 = tf.nn.relu(layer2)

# Output Layer
logits = tf.add(tf.matmul(layer2, weights['out']), biases['out'])
output = tf.nn.softmax(logits)   # probabilities of pass/fail

# ---------------------------
# Define Loss & Optimizer
# ---------------------------
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels=y, logits=logits))
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(loss)

# ---------------------------
# Define Accuracy
# ---------------------------
correct_pred = tf.equal(tf.argmax(output, 1), tf.argmax(y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))

# ---------------------------
# Example Dummy Data
# ---------------------------
# Features: [Age, Hours of Study, Previous Score]
train_X = np.array([[18, 5, 70],
                    [19, 8, 85],
                    [17, 2, 40],
                    [20, 10, 90]], dtype=np.float32)

# Labels: Pass = [1,0], Fail = [0,1]
train_y = np.array([[1,0],
                    [1,0],
                    [0,1],
                    [1,0]], dtype=np.float32)

# ---------------------------
# Training
# ---------------------------
init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)

    for epoch in range(training_epochs):
        _, c = sess.run([optimizer, loss], feed_dict={X: train_X, y: train_y})

        if (epoch+1) % 200 == 0:
            acc = sess.run(accuracy, feed_dict={X: train_X, y: train_y})
            print("Epoch:", epoch+1, "Loss=", c, "Accuracy=", acc)

    print("\nOptimization Finished!")

    # Final prediction
    pred = sess.run(output, feed_dict={X: train_X})
    print("Predicted Probabilities:\n", pred)
```

```python
# Nguyễn Hoàng Tùng

import tensorflow as tf
import numpy as np
from tensorflow.keras.datasets import mnist # Using Keras for dataset loading as it's part of TensorFlow.

# 0. Set up eager execution explicitly for GradientTape (default in TF2.x)
# tf.compat.v1.disable_eager_execution() is often used for graph mode, but we want eager for tape.
# In TF 2.x, eager execution is default, so no explicit call is usually needed.

# 1. Manual Layer and Model Definition
# Implements the logic of a Keras Dense layer manually.
class NaiveDense:
    def __init__(self, input_size, output_size, activation=None):
        self.activation = activation if activation is not None else lambda x: x
        # Initialize weights with small random values.
        w_shape = (input_size, output_size)
        w_initial_value = tf.random.uniform(w_shape, minval=0., maxval=1e-1)
        self.W = tf.Variable(w_initial_value)
        # Initialize biases with zeros.
        b_shape = (output_size,)
        b_initial_value = tf.zeros(b_shape)
        self.b = tf.Variable(b_initial_value)

    def __call__(self, inputs):
        # Forward pass: output = activation(dot(input, W) + b).
        return self.activation(tf.matmul(inputs, self.W) + self.b)

    @property
    def trainable_weights(self):
        # Returns the layer's learnable parameters.
        return [self.W, self.b]

# Manual Sequential model to stack NaiveDense layers.
class NaiveSequential:
    def __init__(self, layers):
        self.layers = layers

    def __call__(self, inputs):
        x = inputs
        for layer in self.layers:
            x = layer(x)
        return x

    @property
    def trainable_weights(self):
        weights = []
        for layer in self.layers:
            weights.extend(layer.trainable_weights)
        return weights

# 2. Data Handling with a BatchGenerator
class BatchGenerator:
    def __init__(self, images, labels, batch_size=128):
        assert len(images) == len(labels)
        self.index = 0
        self.images = images
        self.labels = labels
        self.batch_size = batch_size
        self.num_batches = len(images) // batch_size

    def next(self):
        images = self.images[self.index : self.index + self.batch_size]
        labels = self.labels[self.index : self.index + self.batch_size]
        self.index += self.batch_size
        return images, labels

# Load and prepare the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

train_images = train_images.reshape((60000, 28 * 28)).astype("float32") / 255
test_images = test_images.reshape((10000, 28 * 28)).astype("float32") / 255

# Define the custom model
model = NaiveSequential([
    NaiveDense(input_size=28 * 28, output_size=512, activation=tf.nn.relu),
    NaiveDense(input_size=512, output_size=10, activation=tf.nn.softmax)
])

# 3. Define Loss Function and Optimizer
# Using Keras's built-in sparse_categorical_crossentropy and RMSprop, as they are part of TensorFlow's core.
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy()
optimizer = tf.keras.optimizers.RMSprop(learning_rate=1e-3)
accuracy_metric = tf.keras.metrics.SparseCategoricalAccuracy()

# 4. Custom Training Step
@tf.function # Decorator to compile the function into a TensorFlow graph for speed.
def one_training_step_tf_pure(model, images_batch, labels_batch, loss_fn, optimizer, accuracy_metric):
    with tf.GradientTape() as tape:
        predictions = model(images_batch)
        loss = loss_fn(labels_batch, predictions) # Compute loss.

    # Compute gradients of the loss with respect to the model's trainable weights.
    gradients = tape.gradient(loss, model.trainable_weights)

    # Update weights using the optimizer and computed gradients.
    optimizer.apply_gradients(zip(gradients, model.trainable_weights))

    # Update accuracy metric.
    accuracy_metric.update_state(labels_batch, predictions)
    return loss

# 5. Full Custom Training Loop
print("\nTraining pure TensorFlow model...")
num_epochs = 5
for epoch in range(num_epochs):
    print(f"\nEpoch {epoch + 1}/{num_epochs}")
    batch_generator = BatchGenerator(train_images, train_labels) # Re-initialize for each epoch
    accuracy_metric.reset_state() # Reset metric at the start of each epoch.

    for batch_counter in range(batch_generator.num_batches):
        images_batch, labels_batch = batch_generator.next()
        loss = one_training_step_tf_pure(model, images_batch, labels_batch, loss_fn, optimizer, accuracy_metric)

        if batch_counter % 100 == 0:
            print(f"  Batch {batch_counter}/{batch_generator.num_batches}: Loss={loss.numpy():.4f}, Accuracy={accuracy_metric.result().numpy():.4f}")

    print(f"End of Epoch {epoch + 1}: Training Accuracy={accuracy_metric.result().numpy():.4f}")

# 6. Custom Evaluation Loop
print("\nEvaluating pure TensorFlow model on test data...")
test_batch_generator = BatchGenerator(test_images, test_labels)
test_accuracy_metric = tf.keras.metrics.SparseCategoricalAccuracy()
test_loss_sum = 0.0
total_test_samples = 0

for batch_counter in range(test_batch_generator.num_batches):
    images_batch, labels_batch = test_batch_generator.next()
    predictions = model(images_batch)
    loss = loss_fn(labels_batch, predictions)
    test_loss_sum += loss.numpy() * len(images_batch)
    total_test_samples += len(images_batch)
    test_accuracy_metric.update_state(labels_batch, predictions)

final_test_loss = test_loss_sum / total_test_samples
final_test_accuracy = test_accuracy_metric.result().numpy()
print(f"Pure TensorFlow Test Loss: {final_test_loss:.4f}, Pure TensorFlow Test Accuracy: {final_test_accuracy:.4f}")

# Generate predictions with pure TensorFlow model
print("\nGenerating predictions with pure TensorFlow model...")
# For predictions, we can just call the model with test_images
test_predictions = model(test_images)
predicted_class = np.argmax(test_predictions.numpy())
print(f"First test image predicted class: {predicted_class}")
print(f"First test image true label: {test_labels}")
```

```python
# Nguyễn Hoàng Tùng

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt # For optional plotting

print("--- MNIST Dataset (Multiclass Classification) ---")

# 1. Load the MNIST dataset
# MNIST is a classic dataset for handwritten digit classification (multiclass)
(train_images_mnist, train_labels_mnist), (test_images_mnist, test_labels_mnist) = keras.datasets.mnist.load_data()

# 2. Prepare the data
# Reshape images to 28*28 (784 features) and normalize pixel values to
train_images_mnist = train_images_mnist.reshape((60000, 28 * 28)).astype("float32") / 255
test_images_mnist = test_images_mnist.reshape((10000, 28 * 28)).astype("float32") / 255

# 3. Build the model
# A simple densely connected network is used for multiclass classification.
# The last layer has 10 units (for 10 classes) and a softmax activation for probability distribution.
model_mnist = keras.Sequential([
    layers.Dense(512, activation="relu"),
    layers.Dense(10, activation="softmax")
])

# 4. Compile the model
# Using 'rmsprop' optimizer and 'sparse_categorical_crossentropy' because labels are integers.
model_mnist.compile(optimizer="rmsprop",
                    loss="sparse_categorical_crossentropy",
                    metrics=["accuracy"])

# 5. Train the model
print("Training MNIST model...")
# Training for a few epochs, showing progress.
model_mnist.fit(train_images_mnist, train_labels_mnist, epochs=5, batch_size=128)

# 6. Evaluate the model
print("Evaluating MNIST model...")
test_loss_mnist, test_acc_mnist = model_mnist.evaluate(test_images_mnist, test_labels_mnist)
print(f"MNIST Test Accuracy: {test_acc_mnist:.3f}")


print("\n--- IMDB Dataset (Binary Classification) ---")

# 1. Load the IMDB dataset
# IMDB is a dataset of movie reviews for binary (positive/negative) classification.
# num_words=10000 keeps only the top 10,000 most frequent words.
(train_data_imdb, train_labels_imdb), (test_data_imdb, test_labels_imdb) = keras.datasets.imdb.load_data(num_words=10000)

# 2. Prepare the data
# Convert integer sequences to multi-hot encoded binary vectors (10,000 dimensions).
def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        for j in sequence:
            results[i, j] = 1.
    return results

x_train_imdb = vectorize_sequences(train_data_imdb)
x_test_imdb = vectorize_sequences(test_data_imdb)

# Vectorize labels as float32
y_train_imdb = np.asarray(train_labels_imdb).astype("float32")
y_test_imdb = np.asarray(test_labels_imdb).astype("float32")

# 3. Building your model
# For binary classification, the last layer has 1 unit and 'sigmoid' activation.
model_imdb = keras.Sequential([
    layers.Dense(16, activation="relu"),
    layers.Dense(16, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])

# 4. Compile the model
# Using 'rmsprop' optimizer and 'binary_crossentropy' loss.
model_imdb.compile(optimizer="rmsprop",
                   loss="binary_crossentropy",
                   metrics=["accuracy"])

# 5. Set aside a validation set
# 10,000 samples from the training data are used for validation.
x_val_imdb = x_train_imdb[:10000]
partial_x_train_imdb = x_train_imdb[10000:]
y_val_imdb = y_train_imdb[:10000]
partial_y_train_imdb = y_train_imdb[10000:]

# 6. Train the model
print("Training IMDB model...")
# The validation_data argument is used to monitor performance on unseen data.
history_imdb = model_imdb.fit(partial_x_train_imdb,
                              partial_y_train_imdb,
                              epochs=20, # Training for 20 epochs as in the textbook example
                              batch_size=512,
                              validation_data=(x_val_imdb, y_val_imdb))

# 7. Evaluate the model
print("Evaluating IMDB model...")
# The textbook retrains for 4 epochs on full data after validation. For this example, we evaluate directly.
results_imdb = model_imdb.evaluate(x_test_imdb, y_test_imdb)
loss_imdb, acc_imdb = results_imdb
print(f"IMDB Test Loss: {loss_imdb:.3f}")
print(f"IMDB Test Accuracy: {acc_imdb:.3f}")

# 8. Generate predictions
print("Generating predictions for IMDB test set (first 5 samples):")
predictions_imdb = model_imdb.predict(x_test_imdb[:5])
print(predictions_imdb)
```
