```python
# Nguyễn Hoàng Tùng

# Import thư viện cần thiết
import pandas as pd
import os

# Đường dẫn thư mục chứa dữ liệu
paths = [
    r"D:\lecture\PTIT\20251\IntSys\Viet Nam Stock Market Data",
    r"D:\PTIT\20251\IntSys\Viet Nam Stock Market Data"
]

# Tên các file CSV
files = ["HOSEVNINDEXM30.csv", "HOSEVN100M30.csv", "HOSEVN30M30.csv"]

# Hàm tìm file trong các đường dẫn
def find_file(file_name, search_paths):
    for path in search_paths:
        full_path = os.path.join(path, file_name)
        if os.path.exists(full_path):
            return full_path
    return None

# Load các dataset
datasets = {}
for file in files:
    file_path = find_file(file, paths)
    if file_path:
        datasets[file] = pd.read_csv(file_path)
        print(f"{file} loaded successfully with shape {datasets[file].shape}")
    else:
        print(f"{file} not found in the given paths.")

# Kiểm tra dữ liệu đầu tiên
for name, df in datasets.items():
    print(f"\n{name} head:")
    display(df.head())
```

```python
# Nguyễn Hoàng Tùng

# Import thư viện
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt

# -----------------------------
# Chuẩn bị dữ liệu
# -----------------------------
dfs = []
for name, df in datasets.items():
    df = df.sort_values('time')
    df = df[['close']].copy()
    dfs.append(df)

# Kết hợp dữ liệu theo index
combined_data = pd.concat(dfs, axis=1)
combined_data.columns = ['close1','close2','close3']
combined_data.dropna(inplace=True)

# Chuẩn hóa dữ liệu
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(combined_data)

# -----------------------------
# Tạo dữ liệu chuỗi cho LSTM
# -----------------------------
def create_sequences(data, seq_length=5):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length, 0])  # Dự đoán close1
    return np.array(X), np.array(y)

SEQ_LENGTH = 10  # Sử dụng 10 ngày trước để dự đoán ngày tiếp theo
X, y = create_sequences(scaled_data, SEQ_LENGTH)

# Chia train/test (80%-20%)
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# -----------------------------
# Xây dựng mô hình LSTM >=5 layer
# -----------------------------
model = Sequential([
    LSTM(128, return_sequences=True, input_shape=(SEQ_LENGTH, X_train.shape[2])),
    LSTM(64, return_sequences=True),
    LSTM(32, return_sequences=True),
    LSTM(16, return_sequences=True),
    LSTM(8),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.summary()

# -----------------------------
# Huấn luyện mô hình
# -----------------------------
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.1)

# -----------------------------
# Đánh giá mô hình
# -----------------------------
y_pred = model.predict(X_test)

plt.figure(figsize=(12,6))
plt.plot(y_test, label='Actual')
plt.plot(y_pred, label='Predicted')
plt.title('Dự đoán giá đóng cửa close1 bằng LSTM')
plt.legend()
plt.show()
```

```python
# Nguyễn Hoàng Tùng

# -----------------------------
# Import thư viện
# -----------------------------
import pandas as pd
import numpy as np
import datetime
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# -----------------------------
# Chuẩn bị dữ liệu
# -----------------------------
dfs = []
for name, df in datasets.items():
    df = df.sort_values('time')
    df = df[['close']].copy()
    dfs.append(df)

combined_data = pd.concat(dfs, axis=1)
combined_data.columns = ['close1','close2','close3']
combined_data.dropna(inplace=True)

# Chuẩn hóa dữ liệu
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(combined_data)

SEQ_LENGTH = 10  # Sử dụng 10 ngày gần nhất

# -----------------------------
# Tạo chuỗi dữ liệu cho LSTM
# -----------------------------
def create_sequences(data, seq_length=10):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])  # dự đoán toàn bộ vector [close1, close2, close3]
    return np.array(X), np.array(y)

X, y = create_sequences(scaled_data, SEQ_LENGTH)
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# -----------------------------
# Huấn luyện mô hình LSTM
# -----------------------------
model = Sequential([
    LSTM(128, return_sequences=True, input_shape=(SEQ_LENGTH, X_train.shape[2])),
    LSTM(64, return_sequences=True),
    LSTM(32, return_sequences=True),
    LSTM(16, return_sequences=True),
    LSTM(8),
    Dense(3)  # Dự đoán 3 cổ phiếu cùng lúc
])

model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.1, verbose=0)

# -----------------------------
# Tính RMSE cho các baseline
# -----------------------------
y_true = scaled_data[SEQ_LENGTH:, :]  # Thực tế (close1,2,3)
y_pred_lstm = model.predict(X_test)

rmse_lstm = np.sqrt(mean_squared_error(y_test, y_pred_lstm))
print(f"RMSE LSTM: {rmse_lstm:.5f}")

# 1. Naive
y_naive = scaled_data[SEQ_LENGTH-1:-1, :]
rmse_naive = np.sqrt(mean_squared_error(y_true, y_naive))

# 2. Moving Average
window = 5
y_ma = []
for i in range(SEQ_LENGTH, len(scaled_data)):
    y_ma.append(np.mean(scaled_data[i-window:i, :], axis=0))
y_ma = np.array(y_ma)
rmse_ma = np.sqrt(mean_squared_error(y_true, y_ma))

# 3. Linear Regression
X_lr = scaled_data[SEQ_LENGTH-1:-1, :]
lr_models = [LinearRegression() for _ in range(3)]
y_lr_true = y_true.copy()
y_lr_pred = np.zeros_like(y_lr_true)
for i in range(3):
    lr_models[i].fit(X_lr, y_lr_true[:, i])
    y_lr_pred[:, i] = lr_models[i].predict(X_lr)
rmse_lr = np.sqrt(mean_squared_error(y_lr_true, y_lr_pred))

# -----------------------------
# Chọn mô hình tốt nhất
# -----------------------------
rmse_scores = {
    'Naive': rmse_naive,
    'Moving Average': rmse_ma,
    'Linear Regression': rmse_lr,
    'LSTM': rmse_lstm
}
best_model_name = min(rmse_scores, key=rmse_scores.get)

print("\n=== ĐÁNH GIÁ MÔ HÌNH ===")
for k, v in rmse_scores.items():
    print(f"{k:20s}: RMSE = {v:.5f}")
print(f"\n👉 Mô hình tốt nhất: {best_model_name} (RMSE = {rmse_scores[best_model_name]:.5f})")

# -----------------------------
# Dự báo cho tất cả cổ phiếu
# -----------------------------
last_seq = scaled_data[-SEQ_LENGTH:]
last_seq_input = np.expand_dims(last_seq, axis=0)

if best_model_name == 'LSTM':
    next_scaled = model.predict(last_seq_input)[0]
elif best_model_name == 'Linear Regression':
    next_scaled = np.array([lr_models[i].predict(scaled_data[-1,:].reshape(1,-1))[0] for i in range(3)])
elif best_model_name == 'Moving Average':
    next_scaled = np.mean(scaled_data[-5:, :], axis=0)
else:  # Naive
    next_scaled = scaled_data[-1, :]

# Giải mã về giá thực tế
next_real = scaler.inverse_transform(next_scaled.reshape(1, -1))[0]
today_real = combined_data.iloc[-1].values

# Xác định xu hướng
trends = ['Tăng' if next_real[i] > today_real[i] else 'Giảm' for i in range(3)]

# -----------------------------
# Lấy ngày dự báo
# -----------------------------
last_unix = datasets[list(datasets.keys())[0]]['time'].max()
last_date = datetime.datetime.utcfromtimestamp(last_unix)
next_date = last_date + datetime.timedelta(days=1)

# -----------------------------
# Tạo bảng kết quả
# -----------------------------
results = pd.DataFrame({
    'Cổ phiếu': ['close1', 'close2', 'close3'],
    'Ngày dự báo': [next_date.strftime('%d/%m/%Y')]*3,
    'Giá hiện tại': today_real,
    'Giá dự kiến': next_real,
    'Xu hướng': trends
})

print("\n=== DỰ BÁO TỔNG HỢP ===")
display(results)
```
